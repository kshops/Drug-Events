```python
from bokeh.plotting import figure
from bokeh.io import output_notebook, show, gridplot, output_file
import pandas as pd
import numpy as np
import re
df=pd.read_csv('processed_drug_events_no_age_filt.csv')
df.columns
```

```python
years=np.array([2004,2005,2006,2007,2008,2009,2010,2011,2012,2013,2014])
grouped=df.groupby(['seriousnessdeath','drug0_drugtype','Year'])
events=grouped['drug0_drugtype']
total_events=events.apply(len)

total_events
```

```python
output_file('Figure1b.html')
select_opioid_death=(1,'Opioid Agonist')
select_opioid=(0, 'Opioid Agonist')
opioid_events = total_events[select_opioid_death]+total_events[select_opioid]
opioid_deaths= total_events[select_opioid_death]
select_benzo_death=(1,'Benzodiazepine')
select_benzo=(0, 'Benzodiazepine')
benzo_events = total_events[select_benzo_death]+total_events[select_benzo]
benzo_deaths= total_events[select_benzo_death]
select_anti_death=(1,'Atypical Antipsychotic')
select_anti=(0, 'Atypical Antipsychotic')
anti_events = total_events[select_anti_death]+total_events[select_anti]
anti_deaths= total_events[select_anti_death]
select_nsaid_death=(1,'Nonsteroidal Anti-inflammatory Drug')
select_nsaid=(0, 'Nonsteroidal Anti-inflammatory Drug')
nsaid_events = total_events[select_nsaid_death]+total_events[select_nsaid]
nsaid_deaths= total_events[select_nsaid_death]


p = figure(title="Opioids",width=650, height=550, x_axis_label='Year', y_axis_label='Number of Events', x_range=[2002,2015])


p.quad(left=years-0.4, right=years+0, bottom=0, top=opioid_events, 
     fill_alpha=0.4, line_color='blue', fill_color='blue', legend='Events')
p.quad(left=years-0.4, right=years, bottom=0, top=opioid_deaths, 
    fill_alpha=0.8,line_color='pink', fill_color='pink', legend='Deaths')

p2 = figure(title="Benzodiazepines",width=650, height=550, x_axis_label='Year', y_axis_label='Number of Events')
p2.x_range=p.x_range
p2.y_range=p.y_range

p2.quad(left=years-0.4, right=years+0, bottom=0, top=benzo_events, 
     fill_alpha=0.4, line_color='blue', fill_color='blue', legend='Events')
p2.quad(left=years-0.4, right=years+0, bottom=0, top=benzo_deaths, 
     fill_alpha=0.8,line_color='pink', fill_color='pink', legend='Deaths')

p3 = figure(title="Atypical Antipsychotic",width=650, height=550, x_axis_label='Year', y_axis_label='Number of Events')
p3.x_range=p.x_range
p3.y_range=p.y_range

p3.quad(left=years-0.4, right=years+0, bottom=0, top=anti_events, 
     fill_alpha=0.4, line_color='blue', fill_color='blue', legend='Events')
p3.quad(left=years-0.4, right=years+0, bottom=0, top=anti_deaths, 
     fill_alpha=0.8,line_color='pink', fill_color='pink', legend='Deaths')

p4 = figure(title="NSAID",width=650, height=550, x_axis_label='Year', y_axis_label='Number of Events')
p4.x_range=p.x_range
p4.y_range=p.y_range

p4.quad(left=years-0.4, right=years+0, bottom=0, top=nsaid_events, 
     fill_alpha=0.4, line_color='blue', fill_color='blue', legend='Events')
p4.quad(left=years-0.4, right=years+0, bottom=0, top=nsaid_deaths, 
     fill_alpha=0.8,line_color='pink', fill_color='pink', legend='Deaths')

f = gridplot([[p, p2],[p3,p4]])
show(f)
```

```python
years=np.array([2004,2005,2006,2007,2008,2009,2010,2011,2012,2013,2014])
grouped=df.groupby(['seriousnessdeath','drug0_drugtype','patientsex', 'Year'])
events=grouped['drug0_drugtype']
total_events=events.apply(len)

total_events
```

```python
output_file('Figure1b.html')
select_opioid_m=(1,'Opioid Agonist',1)
select_opioid_f=(1, 'Opioid Agonist', 2)
opioid_m = total_events[select_opioid_m]
opioid_f= total_events[select_opioid_f]
select_benzo_m=(1,'Benzodiazepine',1)
select_benzo_f=(1, 'Benzodiazepine',2)
benzo_m = total_events[select_benzo_m]
benzo_f= total_events[select_benzo_f]
select_anti_m=(1,'Atypical Antipsychotic',1)
select_anti_f=(1, 'Atypical Antipsychotic',2)
anti_m = total_events[select_anti_m]
anti_f= total_events[select_anti_f]
select_nsaid_m=(1,'Nonsteroidal Anti-inflammatory Drug',1)
select_nsaid_f=(1, 'Nonsteroidal Anti-inflammatory Drug',2)
nsaid_m = total_events[select_nsaid_m]
nsaid_f= total_events[select_nsaid_f]


p = figure(title="Opioids",width=650, height=550, x_axis_label='Year', y_axis_label='Reported Deaths', x_range=[2002,2015])


p.quad(left=years-0.4, right=years+0.2, bottom=0, top=opioid_m, 
     fill_alpha=0.4, line_color='blue', fill_color='blue', legend='Male')
p.quad(left=years-0.4, right=years, bottom=opioid_m, top=opioid_m+opioid_f, 
    fill_alpha=0.8,line_color='pink', fill_color='pink', legend='Female')

p2 = figure(title="Benzodiazepines",width=650, height=550, x_axis_label='Year', y_axis_label='Reported Deaths')
p2.x_range=p.x_range
p2.y_range=p.y_range

p2.quad(left=years-0.4, right=years+0.2, bottom=0, top=benzo_m, 
     fill_alpha=0.4, line_color='blue', fill_color='blue', legend='Male')
p2.quad(left=years-0.4, right=years+0, bottom=benzo_m, top=benzo_m+benzo_f, 
     fill_alpha=0.8,line_color='pink', fill_color='pink', legend='Female')

p3 = figure(title="Atypical Antipsychotic",width=650, height=550, x_axis_label='Year', y_axis_label='Reported Deaths')
p3.x_range=p.x_range
p3.y_range=p.y_range

p3.quad(left=years-0.4, right=years+0.2, bottom=0, top=anti_m, 
     fill_alpha=0.4, line_color='blue', fill_color='blue', legend='Male')
p3.quad(left=years-0.4, right=years+0, bottom=anti_m, top=anti_m+anti_f, 
     fill_alpha=0.8,line_color='pink', fill_color='pink', legend='Female')

p4 = figure(title="NSAID",width=650, height=550, x_axis_label='Year', y_axis_label='Reported Deaths')
p4.x_range=p.x_range
p4.y_range=p.y_range

p4.quad(left=years-0.4, right=years+0.2, bottom=0, top=nsaid_m, 
     fill_alpha=0.4, line_color='blue', fill_color='blue', legend='Male')
p4.quad(left=years-0.4, right=years+0, bottom=nsaid_m, top=nsaid_m+nsaid_f, 
     fill_alpha=0.8,line_color='pink', fill_color='pink', legend='Female')

f = gridplot([[p, p2],[p3,p4]])
show(f)
```
